```python
import pandas as pd
import numpy as np
import graphviz as vg
import time
import datetime
import d3py
import pickle
import math
```

```python
xls = pd.ExcelFile("SFMdata.xlsx")
xls1 = xls.parse(0)
xls2 = xls.parse(1)
xls3 = xls.parse(2)
xls1["Organization:ID"] = xls1.index
xls2["ID"] = xls2.index
xls1 = xls1.drop(['Unnamed: 66', 'Unnamed: 67', 'Unnamed: 68', 'Unnamed: 69', 'Unnamed: 70', 'Unnamed: 71', 'Unnamed: 72', 'Unnamed: 73', 'Unnamed: 74', 'Unnamed: 75'], 1)
xls2 = xls2.drop(['Unnamed: 35', 'Unnamed: 36'], 1)
```

```python
def all_indices(value, qlist):
    indices = []
    idx = -1
    while True:
        try:
            idx = qlist.index(value, idx+1)
            indices.append(idx)
        except ValueError:
            break
    return indices

def relationship (typ, inp1, inp2, ask = 1):
    dictionary = {}
    num = int(typ)
    column = int(inp1)
    column2 = int(inp2)
    if num == 1:
        asd = xls1
    elif num == 2:
        asd = xls2
    elif num == 3:
        asd = xls3
    else:
        return ("First number is invalid.")
    a = asd.iloc[:,column].tolist()
    print(a)
    b = asd.iloc[:,column2].tolist()
    if ask == 1:
        start = time.time()
        for i in range(len(a)):
            ert = list()
            part1 = a[i].encode("utf-8", "ignore")
            g = a[i]
            relev = all_indices(g,a)
            for r in relev:
                if type(b[r]) == float:
                    if math.isnan(b[r]):
                        continue
                    else:
                        ert.append(b[r])
                if type(b[r]) != float and pd.isnull(b[r]):
                    continue
                elif type(b[r]) == unicode:
                    s = b[r]
                    ert.append(s)
                else:
                    ert.append(b[r])
            dictionary[part1] = ert
        return dictionary
    elif ask == 2:
        print ("This will find the relationship between " + list(asd)[column] + " and " + list(asd)[column2] + ".")
        answer = raw_input("Is this what you want? (Y or N)")
        if answer == "Y":
            for i in range(len(a)):
                ert = list()
                part1 = a[i].encode("utf-8", "ignore")
                g = a[i]
                relev = all_indices(g,a)
                for r in relev:
                    if type(b[r]) == float:
                        if math.isnan(b[r]):
                            continue
                        else:
                            ert.append(b[r])
                    if type(b[r]) != float and pd.isnull(b[r]):
                        continue
                    elif type(b[r]) == unicode:
                        s = b[r].encode("utf-8", "ignore")
                        ert.append(s)
                    else:
                        ert.append(b[r])
                dictionary[part1] = ert
            return dictionary
        elif answer == "N":
            return -1
        else:
            print("Invalid input.")
            return -1
    else:
        print("Final number is invalid. Must be either 1 or 2.")
        return -1
```

```python
start = time.time()
result = relationship(1,1,56)
f = open("dicts", "w")
pickle.dump(result, f)
f.close()
resultdate = relationship(1,1,62)
f = open("dictsdate", "w")
pickle.dump(resultdate, f)
f.close()
resultpeople1 = relationship(2,8,1)
f = open("dictspeople1", "w")
pickle.dump(resultpeople1, f)
f.close()
resultpeople2 = relationship(2,1,8)
f = open("dictspeople2", "w")
pickle.dump(resultpeople2, f)
f.close()
resultpeople3 = relationship(2,1,11)
f = open("dictspeople3", "w")
pickle.dump(resultpeople3, f)
f.close()
resultlocation = relationship(1,1,44)
f = open("dictslocation", "w")
pickle.dump(resultlocation, f)
f.close()
end = time.time()
total = round((end-start),3)/60
if total < 1:
    print("Completed. This took approximately " + str(total*60) + " seconds.")
else:
    print("Completed. This took approximately " + str(total) + " minutes.")
```
